```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd
import tensorflow as tf
```

```python
# Reads in example csv file to test if working
# Outcome which day will have frost, 1 means there was frost 0 means no frost
File_path = "Resources/clean_weather_data.csv"
weather_df = pd.read_csv(File_path)
weather_df
```

```python
# Separate the Features (X) from the Target (y)
y =weather_df["Outcome(Frost)"]
x =weather_df.drop(columns="Outcome(Frost)")
```

```python
# Split our data into training and testing
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

```python
# Create a Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

```python
# Fit (train) or model using the training data
classifier.fit(X_train, y_train)
```

```python
# Make predictions, 1 means there was frost 0 means no frost
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Accuracy metric
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
